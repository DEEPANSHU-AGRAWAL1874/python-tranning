```python
import seaborn as sns
import numpy as np
```

# there are two functions in seaborn

1. figure level function
2. axis level function


```python
df = sns.load_dataset('penguins')
```

```python
df.sample(10)
```

```python
sns.relplot(data= df,x='flipper_length_mm',y = 'bill_length_mm') # by default it will give scatter plot 
```

```python
sns.relplot(data= df,x='flipper_length_mm',y = 'bill_length_mm',kind = 'line')
```

```python
sns.relplot(data= df,x='flipper_length_mm',y = 'bill_length_mm',hue='species')
```

```python
sns.relplot(data= df,x='flipper_length_mm',y = 'bill_length_mm',style = 'island',hue='species')
```

```python
sns.relplot(data= df,x='flipper_length_mm',y = 'bill_length_mm',size = 'body_mass_g')
```

```python
sns.relplot(data= df,x='flipper_length_mm',y = 'bill_length_mm',hue='species',kind='line',size = 'body_mass_g')
```

```python
sns.lineplot(data = df,x='flipper_length_mm',y = 'bill_length_mm',hue = 'species',style = 'species',size= 'species' )
```

```python
sns.lineplot(data = df,x='flipper_length_mm',y = 'bill_length_mm',style = 'species',size = 'species')
```

```python
sns.relplot(data= df,x='flipper_length_mm',y = 'bill_length_mm',hue = 'sex',style = 'species',col = 'island')
```

```python
sns.relplot(data= df,x='flipper_length_mm',y = 'bill_length_mm',hue = 'sex',style = 'species',row  ='sex')
```

```python
sns.relplot(data= df,x='flipper_length_mm',y = 'bill_length_mm',hue = 'species',style = 'sex',row  ='sex',col = 'island')
```

```python
sns.relplot(data= df,x='flipper_length_mm',y = 'bill_length_mm',col = 'island',col_wrap= 4)
```

## distribution plot


```python
sns.histplot(data=df,x = 'body_mass_g')
```

```python
sns.displot(data = df,x = 'body_mass_g',bins =100)
```

```python
sns.displot(kind = 'hist',data= df, x = 'species')
```

```python
sns.displot(kind = 'hist',data = df, x = 'flipper_length_mm',hue = 'species',element= 'step')
```

```python
sns.displot(kind = 'hist',data = df, x = 'flipper_length_mm',hue = 'species',element= 'poly')
```

```python
sns.displot(kind = 'hist',data = df, x = 'flipper_length_mm',hue = 'species',element= 'bars')
```

```python
sns.displot(kind = 'hist',data = df, x = 'flipper_length_mm', y = 'bill_length_mm')
```

```python
sns.displot(kind = 'hist',data = df, x = 'flipper_length_mm',hue = 'species',element= 'bars',col= 'sex',row = 'species')
```

```python
sns.displot(kind = 'hist',data = df, x = 'flipper_length_mm',hue = 'species',element= 'bars',col= 'sex',col_wrap=2)
```

```python
sns.kdeplot(data= df, x = 'flipper_length_mm')
```

```python
sns.displot(kind = 'kde', data = df, x = 'flipper_length_mm')
```
