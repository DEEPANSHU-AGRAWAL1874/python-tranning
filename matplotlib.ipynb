```python
x= [1,2,3,4,5,6,7,8,9]
y = [2,8,16,10,12,6,12,8,3]
z = [3,12,7,10,17,11,7,13,5]
plt.plot(x,y,color = "#ff0202",linestyle = '--',linewidth= 3,marker = '*',markersize = 10)
plt.plot(x,z,color = "#3c71e5",linestyle = '-',linewidth= 3,marker = 'D',markersize = 10)
plt.title('numbers count frequency')
plt.xlabel('Numbers')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.plot(x,y,label= 'abc')
plt.plot(x,z,label= 'xyz')
plt.title('numbers count frequency')
plt.xlabel('Numbers')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
x = ['python','c','c++','java']
y =[89,62,50,80]
plt.bar(x,y, color = 'yellow', width = 0.4, align='edge',edgecolor = 'r',linewidth =  3, linestyle = '--',alpha =0.5)
plt.title('programing language and popularity')
plt.xlabel('languges')
plt.ylabel('popularity')
plt.show()
```

```python
x = ['python','c','c++','java']
y =[89,62,50,80]
plt.barh(x,y, color = 'orange',  align='edge',edgecolor = 'r',linewidth =  3, linestyle = '--',alpha =0.7)
plt.title('programing language and popularity')
plt.xlabel('languges')
plt.ylabel('popularity')
plt.show()
```

```python
# histogram 
num = np.random.randint(0,50,50)
plt.hist(num,color= 'm',edgecolor= 'black',bins = 15,cumulative= 1)
plt.title('frequency count')
plt.xlabel('count')
plt.ylabel('numbers')
plt.show()
```

```python
num = np.random.randint(0,50,50)
plt.hist(num,histtype= 'step')
plt.title('frequency count')
plt.xlabel('count')
plt.ylabel('numbers')
plt.show()
```

```python
num = np.random.randint(0,50,50)
plt.hist(num,orientation= 'horizantal')
plt.title('frequency count')
plt.xlabel('count')
plt.ylabel('numbers')
plt.show()
```
